```python
# Import Packages

import webbrowser
import speech_recognition as sr
import time
from time import ctime
import playsound 
import os
import random
from gtts import gTTS
```

```python
# Initialize. 

r = sr.Recognizer()

# Input function which takes in sound from microphone.

def record_audio(ask = False):
    with sr.Microphone() as source:
        if ask:
            speak(ask)
        audio = r.listen(source)
        voice_data = ''
        
        # Dealing with potential problems picking up sound. 
        
        try:
            voice_data = r.recognize_google(audio)
            
        except sr.UnknownValueError:
            speak("Can you please Repeat")
            
        except sr.RequestError:
            speak("Services are down at the moment")
            
        return voice_data
    
# Speech function.

def speak(audio_string):
    
    tts = gTTS(text = audio_string, lang = 'en')
    r = random.randint(1,10000000)
    audio_file = 'audio-' + str(r) + '.mp3'
    tts.save(audio_file)
    playsound.playsound(audio_file)
    print(audio_string, )
    os.remove(audio_file)

# Function which gives a simple response to a given input. 
    
def respond(voice_data):
    
    # What is the name of the app. 
    
    if 'what is your name' in voice_data:
        speak("My name is Thomas jr, How may I be of service? ")   
        
    # Asking for the date and time. 
    
    if 'what day and time is it' in voice_data:
        speak(ctime())
        
    # Google Searching. 
    
    if 'look up' in voice_data:
        search = record_audio('What do you want me to look up for you?')
        url = 'https://google.com/search?q=' + search
        webbrowser.get().open(url)
        speak("Here is what I found" + search)
        
    # Finding a location. 
    
    if 'find location' in voice_data:
        location = record_audio('What do you want me to look up for you?')
        url = 'https://google.com/maps/place/' + location + '/&amp;'
        webbrowser.get().open(url)
        speak("Here is what I found for " + location)
        
    # Looking up a Youtube Video. 
    
    if 'video' in voice_data:
        topic = record_audio('What topic should I search?')
        url = 'https://www.youtube.com/results?search_query=' + topic
        webbrowser.get().open(url)
        speak("Here is what I found for " + topic)
        
    # Rating out of ten. 

    if 'rate me' in voice_data:
        speak("Woah did someone drop a dime because you are a dime if i ever saw one")
        
    if 'joke' in voice_data:
        speak('If i had a robot daughter I would let her date you.... just kidding')
        

# Ask user what you can do for them. 


time.sleep(1)        
speak("Hi, How may I help you today? ")
while 1:
    voice_data = record_audio()
    respond(voice_data)
```
